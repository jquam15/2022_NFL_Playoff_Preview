```python
import pandas as pd
import random
import numpy as np
import os
```

```python
passing = pd.read_csv("offensive_passing_stats.csv")
rushing = pd.read_csv("offensive_rushing_stats.csv")
receiving = pd.read_csv("offensive_receiving_stats.csv")
downs = pd.read_csv("offensive_downs.csv")
```

```python
passing = passing.drop(columns = ["20+", "40+", "Lng"])
passing
```

```python
rushing = rushing.drop(columns = ["20+", "40+", "Lng"])
rushing = rushing.rename(columns = {"Att": "Att_Rush", "TD": "Rush_TD"})
```

```python
receiving = receiving.drop(columns = ["Rec", "Yds", "20+", "40+", "Lng", "Rec 1st", "Rec 1st%"])
receiving = receiving.rename(columns = {"TD": "Rec_TD"})
```

```python
downs["3rd_Conv%"] = downs["3rd Md"] / downs["3rd Att"]
downs["4th Conv%"] = downs["4th Md"] / downs["4th Att"]
downs["Team2"] = downs["Team"]
downs = downs.loc[:, "Scrm Plys":]
downs = downs.rename(columns = {"Team2": "Team"})
downs
```

```python
records = pd.read_csv("nfl_records.csv")
records["Win pct"] = (records["W"] + .5*records["T"])/(records["W"] + records["L"] + records["T"])
records
```

```python
offense = pd.merge(passing, rushing, on=["Team", "Year"])
offense = pd.merge(offense, receiving, on=["Team", "Year"])
offense = pd.merge(offense, downs, on=["Team", "Year"])
offense = pd.merge(offense, records, on=["Team", "Year"])
recent_offense = offense[offense["Year"] >= 2000]
recent_offense = recent_offense.reset_index(drop=True)
recent_offense
```

```python
if not os.path.exists("combined_offense.csv"):
    offense.to_csv("combined_offense.csv", index = False)
```

```python
### Defense 
```

```python
d_passing = pd.read_csv("defensive_passing_stats.csv")
d_passing = d_passing.drop(columns = ["20+", "40+", "Lng"])
d_passing
```

```python
d_rushing = pd.read_csv("defensive_rushing_stats.csv")
d_rushing = d_rushing.drop(columns = ["20+", "40+", "Lng", "Rush FUM"])
d_rushing = d_rushing.rename(columns = {"Att": "Att_Rush", "TD": "Rush_TD"})
```

```python
d_receiving = pd.read_csv("defensive_receiving_stats.csv")
d_receiving = d_receiving.drop(columns = ["Rec", "Yds", "20+", "40+", "Lng", "Rec 1st", "Rec 1st%"])
d_receiving = d_receiving.rename(columns = {"TD": "Rec_TD"})
```

```python
d_downs = pd.read_csv("defensive_downs.csv")
d_downs["3rd_Conv%"] = d_downs["3rd Md"] / d_downs["3rd Att"]
d_downs["4th Conv%"] = d_downs["4th Md"] / d_downs["4th Att"]
d_downs["Team2"] = d_downs["Team"]
d_downs = d_downs.loc[:, "Scrm Plys":]
d_downs = d_downs.rename(columns = {"Team2": "Team"})
```

```python
turnovers = pd.read_csv("defensive_turnovers.csv")
turnovers = turnovers.drop(columns = ["FR TD", "INT TD", "Rec FUM", "Rush FUM", "INT Yds", "Lng", "INT"])
```

```python
defense = pd.merge(d_passing, d_rushing, on=["Team", "Year"])
defense = pd.merge(defense, d_receiving, on=["Team", "Year"])
defense = pd.merge(defense, d_downs, on=["Team", "Year"])
defense = pd.merge(defense, turnovers, on=["Team", "Year"])
defense = pd.merge(defense, records, on=["Team", "Year"])
recent_defense = defense[defense["Year"] >= 2000]
recent_defense = recent_defense.reset_index(drop=True)
recent_defense
```

```python
if not os.path.exists("combined_defense.csv"):
    defense.to_csv("combined_defense.csv", index = False)
```
