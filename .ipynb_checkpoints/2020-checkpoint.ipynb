```python
import sklearn
from sklearn.linear_model import LinearRegression
import pandas as pd
```

```python
import matplotlib
from matplotlib import pyplot as plt
import numpy as np
import random
```

```python
corr = pd.read_csv("2020.csv")
corr.head()
```

```python
fig, ax = plt.subplots(figsize = (6,4))
plt.scatter(x=corr["Net Y/A"], y=corr["Win Percentage"])
None
```

```python
lr = LinearRegression()
```

```python
lr.fit(corr[["Net Y/A"]], corr["Win Percentage"])
slope = lr.coef_[0]
intercept = lr.intercept_
slope, intercept
```

```python
fig, ax = plt.subplots(figsize = (10,8))
plt.scatter(x=corr["Net Y/A"], y=corr["Win Percentage"])
x_vals = np.array(ax.get_xlim())
y_vals = intercept + slope * x_vals
plt.plot(x_vals, y_vals, color = 'red')
ax.set_xlabel("Net Yards Per Pass")
ax.set_ylabel("Win Percentage")
for i in range(len(corr)):
    ax.annotate(corr.iloc[i]["Team"], xy = (float(corr.iloc[i]["Net Y/A"]), 
                                            (float(corr.iloc[i]["Win Percentage"]) + random.uniform(0.0, 0.03))))

plt.suptitle("The Effect of Net Yards Per Pass on Win Percentage")
plt.title("2020 NFL Season")
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
None
```

```python
#Correlation coefficient for the above graph
r_squared = lr.score(corr[["Net Y/A"]], corr["Win Percentage"])
r = r_squared ** 0.5
r
```
